```python
def convert_columns_to_float(df):
    for col in df.columns:
        try:
            df[col] = df[col].str.replace(',', '.').astype(float)
        except (ValueError, AttributeError):
            pass
    return df
```

```python
fields = convert_columns_to_float(fields)
value = convert_columns_to_float(value)
works = convert_columns_to_float(works)
```

## Матрицы корреляции

```python
import seaborn as sns

def plot_correlation_heatmap(df, one_hot_columns=None, figsize = (10, 10)):
    df = df.copy()
    if one_hot_columns:
        categorical_df = pd.get_dummies(df[one_hot_columns], drop_first=False).astype(int)
        numeric_df = df.select_dtypes(include=['number'])
        df = pd.concat([numeric_df, categorical_df], axis=1)

    corr_matrix = df.corr()
    plt.figure(figsize=figsize)
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
    plt.title("Матрица корреляции")
    plt.show()

plot_correlation_heatmap(fields, ['region', 'land_type', 'seed_origin',	'technology'], (25, 25))
plot_correlation_heatmap(value, ['inv_item_type'], (7, 7))
plot_correlation_heatmap(works, ['work_type', 'prod_purpose'], (15, 15))
```

## Графички распределений

Единственная проблема - если слишком много NaN, график не строится

```python
def plot_numeric_distributions(df):
    numeric_df = df.select_dtypes(include=['number'])
    num_cols = len(numeric_df.columns)
    fig, axes = plt.subplots(nrows=num_cols, figsize=(8, 4 * num_cols))
    if num_cols == 1:
        axes = [axes]

    for ax, col in zip(axes, numeric_df.columns):
      if col == 'season':
        continue
      sns.histplot(numeric_df[col].dropna(), bins=20, kde=True, ax=ax)
      ax.set_title(f"Распределение {col}")

    plt.tight_layout()
    plt.show()

```

```python
plot_numeric_distributions(fields)
```
