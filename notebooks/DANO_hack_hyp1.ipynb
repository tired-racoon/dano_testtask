## Матрицы корреляции

```python
import seaborn as sns

def plot_correlation_heatmap(df, one_hot_columns=None, figsize = (10, 10)):
    df = df.copy()
    if one_hot_columns:
        categorical_df = pd.get_dummies(df[one_hot_columns], drop_first=False).astype(int)
        numeric_df = df.select_dtypes(include=['number'])
        df = pd.concat([numeric_df, categorical_df], axis=1)
    else:
        df = df.select_dtypes(include=['number'])

    corr_matrix = df.corr()
    plt.figure(figsize=figsize)
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
    plt.title("Матрица корреляции")
    plt.show()


plot_correlation_heatmap(fields, ['region', 'land_type', 'seed_origin',	'technology'], (25, 25))
plot_correlation_heatmap(value, ['inv_item_type'], (7, 7))
plot_correlation_heatmap(works, ['work_type', 'prod_purpose'], (15, 15))
```

### One-hot-encoding для культур урожая

Описание происходящего: отбор сельхозкультур колонок из crop и crop_predecessor с хитом выше threshold (я брал 0.5%)

```python
species = [
    "пшеница", "кукуруза", "люпин", "ячмень", "залежь", "эсп", "рожь", "подс", "сорго",
    "клевер", "кл.", "фестул", "люц", "чечевица", "горох", "соя", "овёс", "гречиха",
    "не используется", "тимофеевка", "фацелия", "тритикале", "райгр", "трав", "рапс",
    "ежа", "строительство", "пастб", "свекла", "не занято", "козлятник", "нут", "луг",
    "аренда", "лён", "загон", "чумиза", "горчица", "боб", "сортоучасток", "смес",
    "рыжик", "площадка", "вика", "кострец", "амарант"
]

species = list(set(species) - {"кл."})

def create_ohe(df, column, species_list):
    # df = df.copy()
    for crop in species_list:
        if crop == "клевер":
            df[f"{column}_клевер"] = df[column].str.contains(r"\b(кл\.|клевер)\b", na=False, regex=True).astype(int)
        else:
            df[f"{column}_{crop}"] = df[column].str.contains(fr"\b{crop}\b", na=False, regex=True).astype(int)
    return df

ohe_fields = create_ohe(final_df, "crop", species)
ohe_fields = create_ohe(ohe_fields, "crop_predecessor", species)

ohe_fields.head()
```

### Гипотеза 1. Смена культур положительно влияет на урожайность

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Предполагается, что исходный DataFrame final_df уже содержит столбцы:
# 'crop', 'crop_predecessor', 'yield_fact'
# Также предполагается, что данные уже очищены от NaN и yield_fact != 0

# Отбираем нужные столбцы и отфильтровываем строки
df_hypot1 = final_df[['crop', 'crop_predecessor', 'yield_fact']].dropna()
df_hypot1 = df_hypot1[df_hypot1['yield_fact'] != 0]

# Задаем список культур для анализа
species = [
    "пшеница", "кукуруза", "люпин", "ячмень", "залежь", "эсп", "рожь", "подс", "сорго",
    "клевер", "фестул", "люц", "чечевица", "горох", "соя", "овёс", "гречиха",
    "не используется", "тимофеевка", "фацелия", "тритикале", "райгр", "трав", "рапс",
    "ежа", "строительство", "пастб", "свекла", "не занято", "козлятник", "нут", "луг",
    "аренда", "лён", "загон", "чумиза", "горчица", "боб", "сортоучасток", "смес",
    "рыжик", "площадка", "вика", "кострец", "амарант"
]
# Исключаем неоднозначные варианты, например "кл."
species = list(set(species) - {"кл."})

results = []

# Для каждой культуры фильтруем данные по текущей культуре и создаем индикатор,
# был ли предыдущим тот же вид культуры
for crop in species:
    # Отбираем строки, где в текущей культуре встречается нужное название
    df_crop = df_hypot1[df_hypot1['crop'].str.contains(fr"\b{crop}\b", case=False, na=False)].copy()

    # Если выборка мала, пропускаем
    if len(df_crop) < 10:
        continue

    # Создаем бинарный индикатор: 1, если в предыдущем сезоне встречается тот же crop, иначе 0
    df_crop['has_predecessor'] = df_crop['crop_predecessor'].str.contains(fr"\b{crop}\b", case=False, na=False).astype(int)

    # Если в выборке представлена только одна группа, пропускаем
    if df_crop['has_predecessor'].nunique() < 2:
        continue

    yield_with = df_crop[df_crop['has_predecessor'] == 1]['yield_fact'].mean()
    yield_without = df_crop[df_crop['has_predecessor'] == 0]['yield_fact'].mean()
    diff = yield_with - yield_without

    results.append({
        'crop': crop,
        'yield_with_predecessor': yield_with,
        'yield_without_predecessor': yield_without,
        'difference': diff,
        'n_with': len(df_crop[df_crop['has_predecessor'] == 1]),
        'n_without': len(df_crop[df_crop['has_predecessor'] == 0])
    })

results_df = pd.DataFrame(results)
print("Результаты анализа:")
print(results_df)

# Визуализация разницы средней урожайности (с предшественником минус без предшественника)
plt.figure(figsize=(14, 7))
sns.barplot(x='crop', y='difference', data=results_df, palette='coolwarm')
plt.xticks(rotation=90)
plt.xlabel('Культура')
plt.ylabel('Разница в урожайности')
plt.title('Разница средней урожайности: с предшественником vs без предшественника')
plt.axhline(0, color='black', linestyle='--')
plt.tight_layout()
plt.show()

# Визуализация средней урожайности по группам для каждой культуры
results_melted = results_df.melt(id_vars='crop',
                                 value_vars=['yield_with_predecessor', 'yield_without_predecessor'],
                                 var_name='group', value_name='yield')

plt.figure(figsize=(14, 7))
sns.barplot(x='crop', y='yield', hue='group', data=results_melted, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Культура')
plt.ylabel('Средняя урожайность')
plt.title('Средняя урожайность с предшественником и без предшественника для каждой культуры')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Предполагается, что исходный DataFrame final_df уже очищен: нет NaN в crop, crop_predecessor и yield_fact, yield_fact != 0
df_hypot1 = final_df[['crop', 'crop_predecessor', 'yield_fact']].dropna()
df_hypot1 = df_hypot1[df_hypot1['yield_fact'] != 0]

# Задаем список культур для анализа
species = [
    "пшеница", "кукуруза", "люпин", "ячмень", "залежь", "эсп", "рожь", "подс", "сорго",
    "клевер", "фестул", "люц", "чечевица", "горох", "соя", "овёс", "гречиха",
    "не используется", "тимофеевка", "фацелия", "тритикале", "райгр", "трав", "рапс",
    "ежа", "строительство", "пастб", "свекла", "не занято", "козлятник", "нут", "луг",
    "аренда", "лён", "загон", "чумиза", "горчица", "боб", "сортоучасток", "смес",
    "рыжик", "площадка", "вика", "кострец", "амарант"
]
# Исключаем неоднозначные варианты
species = list(set(species) - {"кл."})

chi2_results = []

for crop in species:
    # Отбираем строки, где в текущей культуре встречается нужное название
    df_crop = df_hypot1[df_hypot1['crop'].str.contains(fr"\b{crop}\b", case=False, na=False)].copy()

    # Если выборка мала, пропускаем
    if len(df_crop) < 10:
        continue

    # Создаем бинарный индикатор: 1, если в предыдущем сезоне встречается тот же crop, иначе 0
    df_crop['has_predecessor'] = df_crop['crop_predecessor'].str.contains(fr"\b{crop}\b", case=False, na=False).astype(int)

    # Если в выборке представлена только одна группа, пропускаем
    if df_crop['has_predecessor'].nunique() < 2:
        continue

    # Определяем медиану урожайности для данной культуры
    median_yield = df_crop['yield_fact'].median()
    # Создаем бинарную переменную: 1 - высокий урожай, если yield_fact >= медианы, иначе 0
    df_crop['yield_category'] = (df_crop['yield_fact'] >= median_yield).astype(int)

    # Строим таблицу сопряженности:
    # строки: наличие того же предшественника (has_predecessor)
    # столбцы: категория урожайности (0 - низкий, 1 - высокий)
    contingency_table = pd.crosstab(df_crop['has_predecessor'], df_crop['yield_category'])

    # Выполняем χ²-тест
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Рассчитываем долю высоких урожаев для групп
    prop_high_with = contingency_table.loc[1, 1] / contingency_table.loc[1].sum() if 1 in contingency_table.index else None
    prop_high_without = contingency_table.loc[0, 1] / contingency_table.loc[0].sum() if 0 in contingency_table.index else None

    chi2_results.append({
        'crop': crop,
        'chi2_stat': chi2,
        'p_value': p,
        'n_with': len(df_crop[df_crop['has_predecessor'] == 1]),
        'n_without': len(df_crop[df_crop['has_predecessor'] == 0]),
        'prop_high_with': prop_high_with,
        'prop_high_without': prop_high_without
    })
```

```python
chi2_results_df = pd.DataFrame(chi2_results)
print("Результаты χ²-теста для каждой культуры:")
print(chi2_results_df)

# Визуализация: для культур с p-value < 0.05 сравним долю высоких урожаев в группах
significant_crops = chi2_results_df[chi2_results_df['p_value'] < 0.05]
if not significant_crops.empty:
    plt.figure(figsize=(14, 7))
    # Столбчатая диаграмма для доли высоких урожаев
    sns.barplot(x='crop', y='prop_high_with', data=significant_crops, color='blue', label='С предшественником')
    sns.barplot(x='crop', y='prop_high_without', data=significant_crops, color='orange', label='Без предшественника', alpha=0.7)
    plt.xticks(rotation=90)
    plt.xlabel('Культура')
    plt.ylabel('Доля высоких урожаев')
    plt.title('Сравнение доли высоких урожаев с предшественником и без (p < 0.05)')
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("Нет культур с p < 0.05 по χ²-тесту.")
```
